## 1. Naive Implementierung

$X$ sei eine binomialverteilte Zufallsvariable. Implementieren Sie die Funktion *binomial(n, p)*, die - gegeben eine Anzahl von Versuchen $n$ und eine Erfolgswahrscheinlichkeit $p$ - ein $(n+1)$-dimensionales (Numpy-)Array $P$ mit den Wahrscheinlichkeiten $P(X=0), P(X=1), ..., P(X=n)$ zurückgibt.

Implementieren Sie zunächst die Standard-Formel aus den Slides:
$$
P(X=k) = \binom{n}{k} \cdot p^k \cdot (1-p)^{n-k}
$$
Nutzen Sie <a href="https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.special.binom.html">scipy.special.binom()</a> zur Berechnung des Binomialkoeffizienten.
Berechnen Sie $P$ und plotten Sie es mit der unten definierten Methode *barplot()*. Variieren Sie $n=3,10,100$ und $p=0.1, 0.5, 0.99$. Sehen die Werte vernünftig aus?

```python
%matplotlib inline 

from scipy.special import binom
import numpy as np
import matplotlib.pyplot as plt

def binomial(n, p):
    # FIXME: compute array P
    P = []
    for k in range(n+1):
        #print(k)
        binomial = binom(n, k) * pow(p, k) * pow(1-p, n-k)
        print(binomial)
        P.append(binomial)
    return np.array(P)


def barplot(P):
    ind = np.arange(0,len(P))
    fig, ax = plt.subplots()
    rects1 = ax.bar(ind+0.5, P, 0.8, color='red')
    ax.set_ylabel('P(X=k)')
    ax.set_xlabel('k')
    plt.grid(True)
    plt.show()


P = binomial(10, 0.3)
barplot(P)
```

## 3. Rekursionsformel

Implementieren Sie nun eine zweite Methode *binomial_recursive()*, die die Rekursionsformel für die Binomialverteilung verwendet *(siehe Beweis auf dem Übungsblatt)*. 

* Hinweis: Denken Sie daran, P[0] mit $P(X=0)$ zu initialisieren.
* Überprüfen Sie per Barplot, dass Ihr Ergebnis immer noch korrekt ist.
* Testen Sie das Fabrikbeispiel aus Teil 2. Erhalten Sie nun ein plausibles Ergebnis?


```python
import numpy as np

def binomial_recursive(n, p):
    # FIXME: fill P
    # P(X = 0)
    P_list = []
    P = pow(1-p, n-0)
    P_list.append(P)
    for k in range(1, n+1):
        P *= (p / (1-p)) * ((n-k) / (k+1))
        P_list.append(P)
    return np.array(P_list)


P = binomial_recursive(10, 0.5)
barplot(P)


P = binomial_recursive(10000, 0.015)
print(sum(P[151:]))
```
